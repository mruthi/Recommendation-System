```python
import numpy as np
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
rating_df = pd.read_csv("ml-100k/u.data", sep='\t', names=r_cols, encoding='latin-1')
```

```python
rating_df
```

```python
#Read the movie Dataset
i_cols = ['movie_id', 'movie_title' ,'release_date','video_release_date', 'IMDb-URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies_df = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
movies_df
```

```python
#merge two dataframes, movies_df and rating_df, based on the column movie_id. 
#This means that for each matching movie_id in both dataframes, the corresponding rows will be combined.
#add the how parameter to specify the type of join (e.g., inner, left, right, or outer).
merged_data = pd.merge(movies_df, rating_df, on='movie_id',how='left')
```

```python
merged_data
```

```python
final_data = merged_data[["movie_id", "movie_title", "user_id",  "rating"]]
```

```python
final_data
```

```python
#Grouping: ratings.groupby('movie id') groups the ratings DataFrame by the movie id. This means that all rows (ratings) for the same movie id are considered together.
#Counting: ['rating'].count() counts the number of ratings for each movie. It doesn't matter what the actual rating value is; this count represents how many times a movie has been rated.
#Resetting Index: .reset_index() converts the result back into a DataFrame. 
movie_popularity = final_data.groupby('movie_id')['rating'].count().reset_index()
```

```python
movie_popularity
```

```python
movie_popularity.columns = ['movie_id', 'popularity']
```

```python
movie_popularity
```

```python
genre_columns = ['Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama',
                 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
genre_recommendations = {}
for genre in genre_columns:
    # Filter movies in the current genre
    genre_movies = movies_df[movies_df[genre] == 1]
    # Merge with popularity data
    genre_movies = pd.merge(genre_movies, movie_popularity, on='movie_id', how='left')
    # Sort by popularity and select the top 5
    genre_movies = genre_movies.sort_values(by='popularity', ascending=False).head(5)
    # Store the recommendations for the current genre
    genre_recommendations[genre] = genre_movies[['movie_title', 'popularity']]
    # Create a DataFrame to store the recommendations for each genre
    recommendations_df = pd.DataFrame()
#if the column 'movie title' exists in the DataFrame recommendations. If it does, the column's values are added to recommendations_df
for genre, recommendations in genre_recommendations.items():
    recommendations_df[genre] = recommendations['movie_title'].values
    
## Transpose the DataFrame for better readability. each row represents a genre, and columns represent movie titles.   
popular_movies = recommendations_df.T
 
print("The Top 5 movies of each genre are as follows :")
popular_movies
```

```python
# here we are calculating no of times (users) who rated each movies i.e no of users
def get_popular_movies(dataset, top_recos_count):
    popular_movies_grouped_data = dataset.groupby(["movie_title"])["user_id"].count().reset_index()
    popular_movies_grouped_data.rename(columns= {"user_id": "score"}, inplace=True)
    # Here primary sort is Score with descending order and the secondary sort (by movie_title in ascending order) is used 
    # when there are two or more movies with the same score. 
    # This ensures that movies with the same score are ordered alphabetically by their titles, 
    # making the data more organized and predictable.
    popular_movies_sort = popular_movies_grouped_data.sort_values(["score", "movie_title"], ascending=[0,1])

    # This means the ranks will be assigned in descending order of the scores. 
    # In other words, the highest score will get the highest rank (rank 1).
    #This specifies the method used to assign ranks to tied scores. The first method assigns ranks in the order the values 
    #appear in the DataFrame.
    #For example, if two movies have the same score, the movie that appears first in the DataFrame will get the higher rank
    popular_movies_sort["rank"] = popular_movies_sort["score"].rank(ascending=0, method="first")
                                            
    popular_recommendations = popular_movies_sort.head(top_recos_count)
    return popular_recommendations
```

```python
get_popular_movies(final_data, 10)
```
