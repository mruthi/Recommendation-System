```python
import numpy as np
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
rating_df = pd.read_csv("ml-100k/u.data", sep='\t', names=r_cols, encoding='latin-1')
```

```python
rating_df
```

```python
#Read the movie Dataset
i_cols = ['movie_id', 'movie_title' ,'release_date','video_release_date', 'IMDb-URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies_df = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
movies_df 
```

```python
merged_data = pd.merge(movies_df, rating_df, on='movie_id',how='left')
```

```python
merged_data
```

```python
final_data = merged_data[["movie_id", "movie_title", "user_id",  "rating"]]
```

```python
final_data
```

```python
popmovie_group = final_data.groupby("movie_title")['user_id'].count().reset_index()
```

```python
popmovie_group
```

```python
popmovie_group.columns = ['movie_title', 'score']
```

```python
popmovie_group
```

```python
popular_movies_sort = popmovie_group.sort_values(["score", "movie_title"], ascending=False)
```

```python
popular_movies_sort
```

```python
#Rank Calculation: Calculates the rank of each movie based on the score, where the highest score gets rank 1.
#Ranking Method: Uses method="first", which assigns ranks based on the order of appearance in the DataFrame. Movies with the same score get ranks according to their order in the DataFrame.
popular_movies_sort["rank"] = popular_movies_sort["score"].rank(ascending=False, method="first")
```

```python
popular_recommendations = popular_movies_sort.head(10)
```

```python
popular_recommendations
```

```python
def get_popular_movies(final_data, top_recos_count):
    popmovie_group = final_data.groupby("movie_title")['user_id'].count().reset_index()
    popmovie_group.columns = ['movie_title', 'score']
    popular_movies_sort = popmovie_group.sort_values(["score", "movie_title"], ascending=False)
    popular_movies_sort["rank"] = popular_movies_sort["score"].rank(ascending=False, method="first")
    popular_recommendations = popular_movies_sort.head(top_recos_count)
    return popular_recommendations
```

```python
get_popular_movies(final_data, 10)
```
